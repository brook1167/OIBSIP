## Customer and Product Analysis

```python
# Age Distribution

sns.histplot(df['Age'],bins=20)
plt.title('Age Distrubution')
plt.xlabel('Age')
plt.ylabel('Frequencey')
plt.figure(figsize=(12,6))
plt.show()

```

```python
# Gender Distribution

gender_distribution = df['Gender'].value_counts()
gender_distribution.plot(kind='bar', rot=0, title='Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Category Distribution

product_category_distribution = df['Product_Category'].value_counts()

# Visualization
product_category_distribution.plot(kind='bar', rot=45, title='Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
# Heatmap - Correlation Matrix

numeric_columns = df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
